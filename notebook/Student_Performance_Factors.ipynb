**DESCRIPCIÓN DEL DATASET**

**TRATAMIENTO DE VALORES FALTANTES**

Se detectaron valores faltantes en tres columnas: Parental_Education_Level, Teacher_Quality y Distance_from_Home.

Se utilizó la moda para rellenar los valores faltantes de las tres columnas por las siguientes razones:

Las tres columnas son categóricas, por lo cual la media y la mediana no son adecuadas.

Las tres columnas tienen pocos datos faltantes, por lo que se pueden rellenar sin distorsionar demasiado los datos.

Los registros tienen información valiosa en otras columnas y usar la moda permite conservar esos registros.

Estas columnas no son variables críticas como lo serían, por ejemplo, los puntajes de exámenes, en cuyo caso sería preferible eliminar registros.

```python
df['Parental_Education_Level'] = df['Parental_Education_Level'].fillna(df['Parental_Education_Level'].mode()[0])
df['Teacher_Quality'] = df['Teacher_Quality'].fillna(df['Teacher_Quality'].mode()[0])
df['Distance_from_Home'] = df['Distance_from_Home'].fillna(df['Distance_from_Home'].mode()[0])
```

```python
df.isnull().sum()
```

```python
sns.set(style="whitegrid", palette="muted", font_scale=1.1)
```

```python
# 1. Gráfico de dispersión: Horas de estudio semanales vs Puntaje en el examen con color por Nivel de Motivación
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Hours_Studied", y="Exam_Score", hue="Motivation_Level", palette="cool", alpha=0.7)
plt.title("Relación entre horas de estudio y puntaje en el examen según motivación")
plt.xlabel("Horas de estudio semanales")
plt.ylabel("Puntaje en el examen")
plt.legend(title="Motivación")
plt.tight_layout()
plt.show()
```

Este gráfico de dispersión muestra cómo se relacionan las horas de estudio semanales con el puntaje obtenido en el examen, diferenciando los puntos por niveles de motivación (bajo, medio o alto).

Diagnóstico:
En general, se observa una tendencia positiva: a mayor cantidad de horas de estudio, más alto tiende a ser el puntaje en el examen. Esta observación respalda la hipótesis de que el esfuerzo académico tiene un impacto directo sobre el rendimiento.

Sin embargo, al observar la variable Motivation_Level (Alta, Media, Baja), pareciera que los tres niveles están bastante dispersos por todo el gráfico. Es decir, hay estudiantes con alta motivación que tienen puntajes bajos, y estudiantes con baja motivación que tienen puntajes altos. Esto sugiere que la motivación por sí sola no parece tener un peso fuerte ni un patrón claro en el puntaje del examen, al menos teniendo en cuenta esta visual.

```python
# 2. Boxplot: Puntaje en el examen según Motivación
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="Motivation_Level", y="Exam_Score", hue="Motivation_Level", palette="cool", legend=False)
plt.title("Distribución de puntajes por nivel de motivación")
plt.xlabel("Nivel de motivación")
plt.ylabel("Puntaje en el examen")
plt.tight_layout()
plt.show()
```

A partir del gráfico anterior, se hizo un boxplot que muestra solo la relación entre la motivación y el puntaje en el examen, para ver si hay alguna diferencia en las distribuciones de puntajes según el nivel de motivación, sin mezclar otras variables.

El gráfico sugiere que existe una leve tendencia a que estudiantes con mayor motivación obtengan puntajes más altos en el examen, ya que la mediana del grupo con motivación alta es mayor. Sin embargo, la presencia significativa de valores atípicos y la amplia variabilidad dentro de todos los grupos indican que la motivación no es un predictor determinante por sí sola. Por tanto, la hipótesis se sostiene parcialmente, pero se requiere analizarla junto con otras variables (como horas de sueño, calidad docente, etc.) para comprender mejor su impacto en el rendimiento académico.

```python
# 3. Gráfico de dispersión: Relación entre horas de sueño, puntaje en el examen y nivel de motivación
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Sleep_Hours", y="Exam_Score", hue="Motivation_Level", palette="cool", alpha=0.7)
plt.title("Relación entre horas de sueño, puntaje en el examen y nivel de motivación")
plt.xlabel("Horas de sueño")
plt.ylabel("Puntaje en el examen")
plt.legend(title="Nivel de motivación")
plt.tight_layout()
plt.show()
```

Este gráfico muestra la relación entre las horas de sueño y el puntaje en el examen, diferenciando a los estudiantes según su nivel de motivación.

Diagnóstico:
No se observa una clara relación entre las horas de sueño y los puntajes. Estudiantes con diferentes niveles de descanso (desde 4 hasta 10 horas) tienen rendimientos variados.

Los tres niveles de motivación (baja, media, alta) están distribuidos de manera bastante pareja a lo largo de todos los valores de sueño y puntaje. No se nota una agrupación evidente donde, por ejemplo, los estudiantes más motivados tengan consistentemente mejores resultados. Al igual que en los gráficos anteriores, la motivación no parece ser un factor claramente determinante del rendimiento, al menos no por sí sola.

```python
# 4. Boxplot: Puntaje en el examen por Género
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="Gender", y="Exam_Score", hue="Gender", palette="cool", legend=False)
plt.title("Distribución de puntajes por género")
plt.xlabel("Género")
plt.ylabel("Puntaje en el examen")
plt.tight_layout()
plt.show()
```

Este boxplot compara cómo se distribuyen los puntajes del examen según el género de los estudiantes.

Diagnóstico:
A simple vista, las medianas de puntaje para ambos géneros son muy similares, lo que indica que no hay una diferencia significativa en el rendimiento promedio entre hombres y mujeres.

La distribución de los puntajes también es comparable: tanto en hombres como mujeres, los valores están relativamente concentrados en un rango similar, y presentan una cantidad parecida de outliers.

Esta visualización apoya la hipótesis de que no existen diferencias significativas en el rendimiento según el género.

```python
# 5. Gráfico multivariado: Puntaje en el examen según Asistencia, Puntajes Previos y Acceso a Internet
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x="Attendance",
    y="Exam_Score",
    size="Previous_Scores",
    hue="Internet_Access",
    alpha=0.6,
    sizes=(20, 200),
    palette="cool"
)
plt.title("Puntaje en examen según asistencia, puntajes previos y acceso a internet")
plt.xlabel("Asistencia (%)")
plt.ylabel("Puntaje en el examen")
plt.legend(title="Acceso a Internet", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Este gráfico muestra la relación entre el puntaje en el examen y la asistencia de los estudiantes, con un enfoque en el acceso a Internet y los puntajes previos.

Diagnóstico:
Hay una tendencia general a que los estudiantes con mayor asistencia obtienen mejores puntajes. Esto es lógico, ya que los estudiantes que asisten más a clase probablemente tienen un mejor entendimiento del contenido.

Aunque hay una ligera diferencia en los puntajes entre los estudiantes con y sin acceso a Internet, esta diferencia no parece ser la principal variable en la mejora de los puntajes.

Los estudiantes con mejores resultados anteriores tienden a tener un mejor desempeño en el examen actual, lo que indica una consistencia en su rendimiento.

Este gráfico sugiere que la asistencia y los puntajes previos son factores más determinantes en el rendimiento del examen que el acceso a Internet.

```python
# 6. Gráfico de barras: Puntaje en el examen según nivel educativo de los padres
plt.figure(figsize=(10, 6))
avg_scores_by_parent_edu = df.groupby("Parental_Education_Level")["Exam_Score"].mean().sort_values()
sns.barplot(x=avg_scores_by_parent_edu.index, y=avg_scores_by_parent_edu.values, hue=avg_scores_by_parent_edu, palette="cool")
plt.title("Puntaje promedio en el examen según nivel educativo de los padres")
plt.xlabel("Nivel educativo de los padres")
plt.ylabel("Puntaje promedio en el examen")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Este gráfico de barras muestra la relación entre el puntaje promedio en el examen y el nivel educativo de los padres.

Diagnóstico: Se observa que los estudiantes cuyos padres alcanzaron niveles más altos de educación (como posgrado) tienden a obtener mejores puntajes promedio en comparación con aquellos cuyos padres tienen niveles más bajos de educación (como educación secundaria).

Esta tendencia apoya la hipótesis descripta anteriormente en la que se planteó que un mayor nivel educativo de los padres se asocia con un mejor rendimiento de los estudiantes.

La diferencia entre los extremos del gráfico sugiere que el entorno familiar y el nivel educativo podrían estar influyendo en el desempeño escolar del alumnado.

Sería interesante realizar un boxplot para detectar mejor la magnitud de esta diferencia.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="Parental_Education_Level", y="Exam_Score", palette="cool")
plt.title("Distribución de puntajes en el examen según nivel educativo de los padres")
plt.xlabel("Nivel educativo de los padres")
plt.ylabel("Puntaje en el examen")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Este gráfico muestra la distribución de los puntajes en el examen según el nivel educativo de los padres.

La mediana del grupo Postgraduate es un poco más alta que la de College, y ambas son más altas que la del grupo High School. Esto confirma que a medida que el nivel educativo de los padres aumenta, también lo hace el rendimiento promedio del estudiante.

Todos los grupos tienen outliers en la parte superior, lo que indica que hay estudiantes con rendimientos altos en cada categoría. También se observan algunos outliers bajos, especialmente en High School y College, lo que sugiere que los estudiantes con padres con menor nivel educativo tienen más tendencia a puntajes bajos extremos.

Este gráfico visualiza lo que ya se detectó: los estudiantes cuyos padres tienen un nivel educativo más alto tienden a obtener mejores puntajes en los exámenes. Aunque las diferencias no son enormes, sí son consistentes y respaldan la hipótesis de que el nivel educativo de los padres influye en el rendimiento académico.

Por lo que se ve en el gráfico, los puntajes de examen son bastante similares entre estudiantes de escuelas públicas y privadas. Esto sugiere que el tipo de escuela (pública o privada) no parece tener un impacto en los puntajes de examen.

```python
sns.histplot(data=df, x='Physical_Activity', bins=10, kde=True, color='orchid')
plt.title('Distribución de actividad física')
plt.xlabel('Días de actividad física')
plt.ylabel('Frecuencia')
plt.show()
```

Este histograma de distribución de la actividad física indica que la mayoría de los estudiantes hace actividad física entre 2 y 4 días por semana, siendo 3 días lo más común. Muy pocos estudiantes hacen ejercicio todos los días (6-7 días) o no hacen nada (0 días).

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='Physical_Activity', y='Exam_Score', data=df, palette='cool')
plt.title('Relación entre actividad física y puntaje en el examen')
plt.xlabel('Días de actividad física')
plt.ylabel('Puntaje en el examen')
plt.show()
```

El gráfico muestra la distribución de los puntajes en el examen según la cantidad de días a la semana que los estudiantes realizan actividad física.

Se observa que, en general, las distribuciones de puntajes son bastante similares entre los distintos niveles de actividad física. Sin embargo, parecería que hay una pequeña tendencia a que los estudiantes que realizan actividad física más días a la semana tienden a obtener puntajes un poco más altos en el examen. Por otro lado, quienes no realizan actividad física parecerían presentar una mayor dispersión en sus puntajes.

En resumen, aunque la actividad física podría estar relacionada con un mejor desempeño en el examen, la diferencia no es para nada marcada.
